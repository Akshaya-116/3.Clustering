```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
from sklearn.cluster import MeanShift,estimate_bandwidth
```

```python
X,_=make_blobs(n_samples=300,centers=4,cluster_std=0.60,random_state=0)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X[:,0],X[:,1],s=50)
plt.title("Clustering Data")
plt.show()
```

```python
bandwidth=estimate_bandwidth(X,quantile=0.2,n_samples=300)
```

```python
bandwidth
```

```python
X
```

```python
ms=MeanShift(bandwidth=bandwidth)
ms.fit(X)
```

```python
labels=ms.labels_
cluster_centers=ms.cluster_centers_
n_clusters=len(np.unique(labels))
```

```python
labels
```

```python
plt.figure(figsize=(8, 6))
palette = sns.color_palette("bright", n_clusters)

for k, col in zip(range(n_clusters), palette):
    members = labels == k
    plt.scatter(X[members, 0], X[members, 1], s=50, c=[col], label=f"Cluster {k}")

plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], s=200, c='black', marker='X', label='Centroids')
plt.title("Mean-Shift Clustering Results")
plt.legend()
plt.show()
```
