```python
# Importing libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
# Selecting required features
X = dataset.iloc[:, [3, 4]].values
```

```python
# Fitting Affinity Propagation to the dataset
from sklearn.cluster import AffinityPropagation
ap = AffinityPropagation(random_state=0)
y_ap = ap.fit_predict(X)
```

```python
# Visualising the clusters (automatic loop method)
plt.figure(figsize=(8,6))
for i in np.unique(y_ap):
    plt.scatter(X[y_ap == i, 0], X[y_ap == i, 1], s=100, label=f'Cluster {i}')

plt.title('Clusters of customers (Affinity Propagation)')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Optional: Print number of clusters
print('Number of clusters formed:', len(np.unique(y_ap)))
```
